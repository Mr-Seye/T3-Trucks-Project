```python
import os
import awswrangler as wr
import boto3
from dotenv import load_dotenv

load_dotenv()

session = boto3.Session(region_name=os.environ["AWS_REGION"])
db = os.environ["ATHENA_DATABASE"]
table = os.environ.get("ATHENA_TABLE_TRANSACTIONS", "transaction")
```

```python
# 1) S3 location of the table
loc = wr.catalog.get_table_location(
    database=db, table=table, boto3_session=session)
print("TABLE LOCATION:", loc)
```

```python
# 2) Table definition (includes partition keys, columns, etc.)
tbl = wr.catalog.get_table(database=db, table=table, boto3_session=session)
print("PARTITION KEYS:", [p["Name"] for p in tbl.get("PartitionKeys", [])])
print("COLUMNS:", [c["Name"]
      for c in tbl.get("StorageDescriptor", {}).get("Columns", [])])
```
